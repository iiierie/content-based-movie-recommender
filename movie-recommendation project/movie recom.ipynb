```python
##now we need to fetch only the genres name from the genres string of 'list of dictionaries' 
## we use ast.literal_eval to convert the string of list to list 
## getname function returns the return_list after appending only the genre names from each movie genres
import ast
def getname(obj):
    return_list = []
    for i in ast.literal_eval(obj):
        return_list.append(i['name'])
    return return_list ## list containing only the genre names
```

```python
movies['genres'] =  movies['genres'].apply(getname) ## updating the genres to list containing only the genres names
```

```python
##function to extract only 3 names and append it to list L and return L
def getnames2(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if(count !=3):
            L.append(i['name'])
            count+=1
        else:
            break
    return L
```

## since tags contains redundant words like "accept", "accepted","acceptance" so we need to change such words into their stem word "accept" using porter stemer

```python
from nltk import PorterStemmer
st = PorterStemmer() ##st.stem(strings to stem)
```

```python
def stemfunction(text):
    L = []
    for i in text.split():
        L.append(st.stem(i))
    return " ".join(L)
```

```python
st.stem("loved") ##example
```

```python
st.stem("racing") ##example
```

```python
newdf = newdf.reset_index(drop=True)
newdf['tags']=newdf['tags'].apply(stemfunction)
```

## next we need to vectorize our data using countvectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(stop_words="english",max_features= 5000)
## sabai movie ko tags lai merge grxa ani picks 500 most repeating features ani individual movie ma kun kun tag kati patak ako xa 500 size ko arrray ma store grdinxa
```

## certain movie ko vector representation in 500x500 plane ma kun chai najik xa is given by smallest cosine angle to it. so cosine similarity nikalna lako so that most similar movie recommend hos

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
np.set_printoptions(threshold = 100)
similarity ## movies ko aru harek movie snga ko similarity
```

```python
np.set_printoptions(threshold = np.inf)
similarity[0] ##first movie ko similarity with other 3965 movies in order
```

```python
sorted(list(enumerate(similarity[0])), reverse = True , key = lambda x:x[1]) ##index giving so that orignal movie order doesnt get lost while sorting
```

```python
newdf[newdf['title']=='Avatar'].index[0]
```

```python
def recommend(movie):
    ## yo movie ko index find grne suru ma 
    movie_index = newdf[newdf['title']== movie].index[0] ## new dataframe only containing the index of that particluar movie
    ## then find the similiar vector for this particular movie
    curr_movie_similarity = similarity[movie_index] ## similarity array of that movie to see which movie is similiar to that 
    movies_list = sorted(list(enumerate(curr_movie_similarity)), reverse = True , key = lambda x:x[1])[1:6] ##sorting so that highest similarity vako movie suru ma aos and then slicing is 1 to 6 i.e top 3
    
    for i in movies_list:
        print(newdf.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
